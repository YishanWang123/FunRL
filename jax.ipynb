```python
import jax
import jax.numpy as jnp
import flax
import flax.linen as nn
import optax
from flax.training import train_state
from flax.training.train_state import TrainState


key = jax.random.PRNGKey(0)
q_key, x_key = jax.random.split(key)

x = jax.random.uniform(x_key, (1, 1,1,1))
class QNet(nn.Module):
    @nn.compact
    def __call__(self,x):
        x = jnp.transpose(x, (0, 3, 1, 2))  # NHWC to NCHW
        x = x/(225.0)
        x = nn.Conv(features=32, kernel_size=(8, 8), strides=(4, 4), use_bias=False)(x)
        x = nn.relu(x)
        x = nn.Conv(features=64, kernel_size=(4, 4), strides=(2, 2), use_bias=False)(x)
        x = nn.relu(x)
        x = nn.Conv(features=64, kernel_size=(3, 3), strides=(1, 1), use_bias=False)(x)
        x = nn.relu(x)
        x = x.reshape((x.shape[0], -1))  # Flatten
        x = nn.Dense(features=512, use_bias=False)(x)
        x = nn.relu(x)
        x = nn.Dense(features=4, use_bias=False)(x)  # Assuming 4 actions
        return x
class TrainState(TrainState):
        target_params: flax.core.FrozenDict

# def QNet_apply(params, x):
#     return QNet().apply({'params': params}, x)

params = QNet().init(q_key,x)
q_state = TrainState.create(
     apply_fn = QNet.apply,
     params = params,
     target_params = params,
     tx = optax.adam(learning_rate=0.0001),
#      tx_state = optax.adam(learning_rate=0.0001).init(params)
     )


q_state
    
    
```
